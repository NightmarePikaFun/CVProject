```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image, ImageDraw
```

```python
pic = cv.imread('Sugar.jpg')
```

```python
plt.imshow(pic)
```

```python
img = cv.cvtColor(pic,cv.COLOR_BGR2GRAY)
plt.imshow(img, cmap ="gray")
```

```python
for i in range(len(img)):
  for j in range(len(img[i])):
    if(img[i][j]<180):
      img[i][j]=0
    else:
      img[i][j]=255

plt.imshow(img)
```

```python
fg = cv.erode(img,None,iterations = 2)
fg = cv.dilate(fg,None,iterations = 1)

contours1,_ = cv.findContours(fg, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
```

```python
imgCopy = pic.copy()
sugar = 0
spoon = 0
for i in range(len(contours1)):
  if (len(contours1[i])<36):
    cv.drawContours(imgCopy, contours1, i, (0,0,255), 1)
    sugar+=1
  else:
    cv.drawContours(imgCopy, contours1, i, (255,0,0), 1)
    spoon+=1

plt.imshow(imgCopy)
print(' Сахар:',sugar,'\n','Ложки:',spoon)
```
